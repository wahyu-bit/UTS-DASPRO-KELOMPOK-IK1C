```python
print("            DEPOT THAI TANIC           ")
print("     Jalan aja jadian nggak No.34      ")
print("---------------------------------------")
print("Kode         Menu                Harga ")
print("O      Thai Tea Original          8000 ")
print("A      Thai Tea Avocado          10000 ")
print("R      Thai Tea Red Velvet       10000 ")
print("C      Thai Tea Chocolate        10000 ")
print("T      Thai Tea Taro             10000 ")
print("P      Thai Tea Peanut           10000 ")
print("S      Thai Tea Strawberry       10000 ")
print("B      Thai Tea Buble Gum        10000 ")
print("M      Thai Tea Moccacino        10000 ")
print("D      Thai Tea Durian           12000 ")
print("---------------------------------------")
kode = []
jenis_menu=[]
harga=[]
jumlah=[]
jumlah_menu = []
banyak_menu = int(input("Banyak Menu : "))
i=0
while i < banyak_menu :
    print("Menu ke- ",i+1)
    kode.append(input("Kode Menu [O/A/R/C/T/P/S/B/M/D] :"))
    jumlah_menu.append(int(input("Jumlah Menu : ")))
    if kode[i] == "O" or kode[i]== "o":
        jenis_menu.append("Thai Tea Original")
        harga.append("8000")
        jumlah.append(banyak_menu[i] * int("8000"))
    elif kode[i] == "A" or kode[i]== "a":
        jenis_menu.append("Thai Tea Avocado")
        harga.append("10000")
        jumlah.append(banyak_menu[i] * int("10000"))
    elif kode[i] == "R" or kode[i]== "r":
        jenis_menu.append("Thai Tea Red Velvet")
        harga.append("10000")
        jumlah.append(banyak_menu[i] * int("10000"))
    elif kode[i] == "C" or kode[i]== "c":
        jenis_menu.append("Thai Tea Chocolate")
        harga.append("10000")
        jumlah.append(banyak_menu[i] * int("10000"))
    elif kode[i] == "T" or kode[i]== "t":
        jenis_menu.append("Thai Tea Taro")
        harga.append("10000")
        jumlah.append(banyak_menu[i] * int("10000"))
    elif kode[i] == "P" or kode[i]== "p":
        jenis_menu.append("Thai Tea Peanut")
        harga.append("10000")
        jumlah.append(banyak_menu[i] * int("10000"))
    elif kode[i] == "S" or kode[i]== "s":
        jenis_menu.append("Thai Tea Strawberry")
        harga.append("10000")
        jumlah.append(banyak_menu[i] * int("10000"))
    elif kode[i] == "B" or kode[i]== "b":
        jenis_menu.append("Thai Tea Bubble Gum")
        harga.append("10000")
        jumlah.append(banyak_menu[i] * int("10000"))
    elif kode[i] == "M" or kode[i]== "m":
        jenis_menu.append("Thai Tea Moccacino")
        harga.append("10000")
        jumlah.append(banyak_menu[i] * int("10000"))
    elif kode[i] == "D" or kode[i]== "d":
        jenis_menu.append("Thai Tea Durian")
        harga.append("12000")
        jumlah.append(banyak_menu[i] * int("12000"))
    else:
        jenis_menu.append("Kode Salah")
        harga.append("0")
        jumlah.append(banyak_menu[i] * int("0"))  
    i=i+1
    
print("                          DEPOT THAI TANIC                     ")
print("                   Jalan aja jadian nggak No.34                ")
print("---------------------------------------------------------------")
print("No       Menu                 Harga        Banyak        Jumlah")
print("                              Satuan       Pesanan       Harga ")
print("---------------------------------------------------------------")

jumlah_bayar = 0
a=0
while a < banyak_menu :
    jumlah_bayar = jumlah_bayar + jumlah[a]
    print("%i     %s      %s       %i        %i" %(a+1, jenis_menu[a], harga[a], jumlah_menu[a], jumlah[a]))
    a = a + 1
    
jumlah_bayar=int(input("                                Jumlah Bayar  Rp. "))
pajak = jumlah_bayar * 0.1
total_bayar = jumlah_bayar + pajak
print("\n")
print("                                Pajak 10 % Rp. ",pajak)
print("                                Total Bayar Rp. ",total_bayar)
```
